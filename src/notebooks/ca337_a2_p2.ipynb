```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(test_data)

X_train, X_test, y_train, y_test = train_test_split(X, test_data_labels, test_size=0.75, random_state=42)

dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)

lr_classifier = LogisticRegression()
lr_classifier.fit(X_train, y_train)

y_pred_dt = dt_classifier.predict(X_test)
y_pred_lr = lr_classifier.predict(X_test)

accuracy_dt = accuracy_score(y_test, y_pred_dt)
accuracy_lr = accuracy_score(y_test, y_pred_lr)

print("Decision Tree Classifier Accuracy:", accuracy_dt)
print("Logistic Regression Accuracy:", accuracy_lr)
print(y_pred_dt)
```

```python
from sklearn.metrics import confusion_matrix

correct_both = []
incorrect_both = []
correct_dt_incorrect_lr = []
correct_lr_incorrect_dt = []

for i in range(len(y_pred_dt)):
    label = y_test[i]
    prediction_dt = y_pred_dt[i]
    prediction_lr = y_pred_lr[i]
    text = test_data[i]

    if label == prediction_dt and label == prediction_lr:
        correct_both.append((text, label))
    elif label != prediction_dt and label != prediction_lr:
        incorrect_both.append((text, label, prediction_dt, prediction_lr))
    elif label == prediction_dt and label != prediction_lr:
        correct_dt_incorrect_lr.append((text, label, prediction_lr))
    elif label == prediction_lr and label != prediction_dt:
        correct_lr_incorrect_dt.append((text, label, prediction_dt))

print("Decision Tree Classifier Accuracy:", accuracy_dt)
print("Confusion Matrix for Decision Tree Classifier:")
print(confusion_matrix(y_test, y_pred_dt))
print("\nLogistic Regression Accuracy:", accuracy_lr)
print("Confusion Matrix for Logistic Regression:")
print(confusion_matrix(y_test, y_pred_lr))

print("\nCases that both models classified correctly:")
for text, label in correct_both:
    print(f"Text: {text}, Actual Label: {label}")

print("\nCases that both models classified incorrectly:")
for text, actual, dt_prediction, lr_prediction in incorrect_both:
    print(f"Text: {text}, Actual Label: {actual}, DT Prediction: {dt_prediction}, LR Prediction: {lr_prediction}")

print("\nCases that were correctly classified by Decision Tree and incorrectly by Logistic Regression:")
for text, actual, lr_prediction in correct_dt_incorrect_lr:
    print(f"Text: {text}, Actual Label: {actual}, LR Prediction: {lr_prediction}")

print("\nCases that were correctly classified by Logistic Regression and incorrectly by Decision Tree:")
for text, actual, dt_prediction in correct_lr_incorrect_dt:
    print(f"Text: {text}, Actual Label: {actual}, DT Prediction: {dt_prediction}")

```
